### 2.2. Valores Ausentes em `u_mean` e `T`

- Existem 10 valores ausentes em `u_mean` e 2 em `T`. As demais variáveis estão completas (sem nulos).
- Recomenda-se tratar esses valores antes do treinamento: A. Imputação ou B. Remoção das linhas.
- Se `u_mean` não estiver fortemente correlacionado com outras variáveis, pode-se usar a média ou mediana.


```python
# criar uma matrix de correlação to evaluate the possible imputation strategies
corr_matrix = train_df.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap with the mask
sns.heatmap(
    corr_matrix,
    mask=mask,
    cmap='coolwarm',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    square=True,
    cbar_kws={"shrink": 0.8}
)

plt.title('Correlation Matrix', fontsize=14)
plt.tight_layout()
plt.show()
```

- `u_mean` apresenta forte correlação negativa, enquanto `T` mostra forte correlação positiva com `conversion`.
- Como os dados ausentes são <1% e ambas as variáveis são cruciais, a imputação é recomendada.
- Para `T`, é possível imputar por regressão, dado que está correlacionada com outras variáveis como `flow_rate` e `radius`.


```python
# Impute missing values in T using regression with flow_rate and radius

# Step 1: Define the mask *before* imputation
missing_mask = train_df['T'].isnull()

# Step 2: Separate the data
known = train_df[~missing_mask]
unknown = train_df[missing_mask]

reg = LinearRegression()
reg.fit(known[['flow_rate', 'radius']], known['T'])

# Step 4: Impute only the missing values
train_df.loc[missing_mask, 'T'] = reg.predict(unknown[['flow_rate', 'radius']])

# Step 5: Plot with preserved mask
plt.figure(figsize=(10, 2))
sns.stripplot(x=train_df['T'], color='gray', alpha=0.3)
sns.stripplot(x=train_df.loc[missing_mask, 'T'], color='red', size=8, label='Imputed Points')
plt.title('T Values with Highlighted Imputed Points')
plt.legend()
plt.show()

# Optional: Check values
print("Imputed values:\n", train_df.loc[missing_mask, 'T'].values)
```

> Para `u_mean`, pode-se reconstruir os valores com base na relação física:  
> `u_mean = flow_rate / area`.  
> Esta imputação respeita os princípios da dinâmica dos fluidos e é mais confiável do que imputações via regressão, que podem introduzir viés se a relação física já for conhecida.


```python
# Impute missing values in u_mean using flow_rate and area (u_mean = flow_rate / area)

# First, ensure 'area' exists (area = pi * radius^2)
if 'area' not in train_df.columns:
    train_df['area'] = np.pi * train_df['radius']**2

# Step 1: Define the mask for missing u_mean
missing_mask = train_df['u_mean'].isnull()

# Step 2: Calculate u_mean for missing values
# (flow_rate and area must not be null)
calc_mask = missing_mask & train_df['flow_rate'].notnull() & train_df['area'].notnull()
train_df.loc[calc_mask, 'u_mean'] = train_df.loc[calc_mask, 'flow_rate'] / train_df.loc[calc_mask, 'area']

# Optional: Print imputed values
print("Imputed u_mean values (from flow_rate/area):\n", train_df.loc[calc_mask, 'u_mean'].values)

# If any u_mean values are still missing, you may want to impute them with another method (e.g., median)
if train_df['u_mean'].isnull().sum() > 0:
    print("Warning: Some u_mean values are still missing after calculation. Consider further imputation.")
```

```python
# ------------------------------------------------------------------
# 3. Loop over models, tune, cross-validate and evaluate on test set
# ------------------------------------------------------------------
trained_models = {}  # model name → trained GridSearchCV object

n_splits = 5      # number of folds used in CV

cv = KFold(n_splits=n_splits, shuffle=True, random_state = random_state)
scorers = {
    "RMSE": "neg_root_mean_squared_error",
    "MSE": "neg_mean_squared_error",
    "MAE": "neg_mean_absolute_error",
    "Explained Variance": "explained_variance",
    "R2": "r2"
}

cv_results, test_results = {}, {}

print(f"Hyperparameter tunning via grid search and using K-fold cross validation with {n_splits} folds")

for name, (estimator, grid) in models.items():
    print(f"\n*** {name} model ***\n    ", end = '')

    pipe = Pipeline(steps=[("prep", preprocess), ("model", estimator)])

    search = GridSearchCV(
        pipe,
        param_grid=grid,
        scoring="neg_root_mean_squared_error",
        cv=cv,
        n_jobs=-1,
        verbose = 2,
    )
    search.fit(X_train, y_train)
    trained_models[name] = search  # Store the GridSearchCV object

    if len(search.best_params_) > 0:
        print("    Best hyper-parameters found:")
        for key, value in search.best_params_.items():
            print(f'      + {key.replace("model__", "")}: {value}')

    print(f'    Metrics computed using CV:')
    # Cross-validation metrics for the best hyperparameters
    cv_metrics = cross_validate(
        search.best_estimator_,
        X_train,
        y_train,
        cv=cv,
        scoring=scorers,
        n_jobs= -1,
        return_train_score = True,
        verbose = 0
    )

    cv_metrics = pd.DataFrame(cv_metrics)

    neg_metrics = {"RMSE", "MSE", "MAE"}        # those that came from "neg_*" scorers

    def pos(arr, metric):        # flip sign for the neg_metrics (minimize)
        return -arr if metric in neg_metrics else arr

    summary = {
        m: (pos(cv_metrics[f"test_{m}"], m).mean(),
            pos(cv_metrics[f"test_{m}"], m).std())
        for m in ["RMSE", "MSE", "MAE", "Explained Variance", "R2"]
    }
    
    cv_results[name] = summary
    
    for m, (mean, std) in summary.items():
        metric_name = (m + 20*'.')[0:20]
        print(f"      {metric_name:20s}: {mean:7.4f} ± {std:7.4f}")

    # ── Test-set evaluation ────────────────────────────────────────
    y_pred = search.best_estimator_.predict(X_test)

    test_results[name] = {
    "RMSE": np.sqrt(mean_squared_error(y_test, y_pred)),
    "MSE":  mean_squared_error(y_test, y_pred),
    "MAE":  mean_absolute_error(y_test, y_pred),
    "Explained Variance": explained_variance_score(y_test, y_pred),
    "R2":   r2_score(y_test, y_pred),
    "y_pred": y_pred,
    }

# ------------------------------------------------------------------
# 4. Pick the overall best algorithm (lowest mean CV-RMSE)
# ------------------------------------------------------------------

best_model_name = min(cv_results, key=lambda k: cv_results[k]["RMSE"][0])

# ------------------------------------------------------------------
# 5. Print the metrics
# ------------------------------------------------------------------

print(f"\n{65*'='}\n\nSelected best model by CV-RMSE: ***{best_model_name}***")

print(f"\n{65*'='}\n\nMetrics computed on the unseen test-set:")
for name, metrics in test_results.items():
    print(f"\n{name}:")
    for m in ["RMSE", "MSE", "MAE", "Explained Variance", "R2"]:
        metric_name = (m + 20*'.')[0:20]
        print(f"  {metric_name:20s}: {metrics[m]:7.4f}")
```

> **Note:** In a real-world deployment (where we only have features, not true *y*), we wouldn’t generate all the following plots for unseen data, because we don’t know the true outputs.   
> Instead, we should apply only the single best model to new inputs and deploy its predictions.  
> Here we are comparing all models (with their best hyperparameters determined via grid search) with the *known* test data, just to show the predictive capabilities of each model on previously unseen data.

```python
# ------------------------------------------------------------------
# 6. Scatter plots: ŷ vs y  (test set)
# ------------------------------------------------------------------
print(f"\n{70*'='}\n{10*' '}Predictions vs. True values on the unseen Test-set:\n{70*'='}\n")

n_models = len(test_results)
n_cols   = 2
n_rows   = (n_models+n_cols-1) // n_cols

fig, axes = plt.subplots(
    n_rows, n_cols,
    figsize=(4 * n_cols, 4 * n_rows),
    squeeze=False         # always 2-D array
)

for ax, (name, res) in zip(axes.ravel(), test_results.items()):
    # scatter diagonal reference line
    ax.scatter(y_test, res["y_pred"], alpha=0.6)
    lims = [min(y_test.min(), res["y_pred"].min()),
            max(y_test.max(), res["y_pred"].max())]
    ax.plot(lims, lims, linestyle="--")

    # metrics panel
    metrics_text = (
        f"RMSE.....: {res['RMSE']:7.4f}\n"
        f"MSE......: {res['MSE']:7.4f}\n"
        f"MAE......: {res['MAE']:7.4f}\n"
        f"Expl.Var.: {res['Explained Variance']:7.4f}\n"
        f"R².......: {res['R2']:7.4f}"
    )
    ax.text(
        0.05, 0.95, metrics_text,
        transform=ax.transAxes,
        fontsize=10, fontfamily="monospace",
        verticalalignment="top",
        bbox=dict(boxstyle="round,pad=0.5",
                  facecolor="white", alpha=0.8)
    )

    # labels & layout
    ax.set(
        title=f"{name}: $\hat{{y}}$ vs $y$ (test)",
        xlabel="True y",
        ylabel="Predicted y"
    )

# Hide any unused subplot slots (e.g. when #models is odd)
for ax in axes.ravel()[len(test_results):]:
    ax.set_visible(False)

plt.tight_layout()
plt.show()
```

Although Gradient Boosting showed the best performance during cross-validation (CV), its error increased noticeably on the unseen test set, suggesting possible overfitting to the training folds. In contrast, the FCNN maintained consistent performance between CV and test, indicating stronger generalization. This difference highlights that CV is an estimate, and the true test set remains the most reliable benchmark for final model evaluation.

## 7. Modelo Final e Previsão no Conjunto de Teste

Prever os valores de conversão para o conjunto **Test_data_reactor_2.csv** usando o melhor modelo identificado.

- Reajustar o modelo vencedor usando todo o conjunto de treino.
- Usá-lo para prever a conversão no conjunto de teste externo.

```python
# 1. Load the external test dataset
external_test_df = pd.read_csv("Test_data_reactor_2.csv")

print("Test size:", external_test_df.shape)

# 2. Define the same feature engineering steps used in training
def apply_feature_engineering(df):
    # Base physical features
    df['volume'] = np.pi * df['radius']**2 * df['length']
    df['residence_time'] = df['length'] / df['u_mean']
    # Optional interactions, use if trained model includes them
    # df['c0_T'] = df['c0'] * df['T']
    df['mass_flux_density'] = df['c0'] * df['u_mean']
    
    # Log-transformed features
    df['log_residence_time'] = np.log1p(df['residence_time'])
    df['log_mass_flux_density'] = np.log1p(df['mass_flux_density'])
    
    return df

# 3. Apply feature engineering
external_test_df = apply_feature_engineering(external_test_df)

# 4. Check for missing or mismatched columns
missing_cols = set(X_train.columns) - set(external_test_df.columns)
if missing_cols:
    print("❌ Missing columns in external test set:", missing_cols)
else:
    print("✅ All required columns are present.")

# 5. Subset the features to match X_train
X_external_test = external_test_df[X_train.columns].copy()

# 6. Check and optionally handle missing values in test set
missing_summary = X_external_test.isnull().sum()
missing_in_features = missing_summary[missing_summary > 0]

if not missing_in_features.empty:
    print("⚠️ Missing values found. Applying median imputation to:")
    print(missing_in_features)
    X_external_test = X_external_test.fillna(X_external_test.median())
else:
    print("✅ No missing values. Proceeding to prediction.")
```

## 8. Resultados e conclusões



- O modelo selecionado pelo pipeline através da validação cruzada foi o **Gradient Boosting**, enquanto o **FCNN (Fully Connected Neural Network)** apresentou o melhor desempenho no conjunto de teste. Essa diferença sugere que a distribuição ou estrutura dos dados de teste favoreceu a generalização do FCNN, enquanto o Gradient Boosting pode ter se ajustado demais às dobras da validação.
  
- A imputação de `u_mean` utilizando a relação física `u_mean = flow_rate / area` garante consistência com os princípios da dinâmica dos fluidos e resulta em valores mais confiáveis e interpretáveis do que a imputação via regressão.

- A variação entre K=5 e K=10 na validação cruzada não alterou significativamente o modelo selecionado nem a ordem de desempenho dos algoritmos — o Gradient Boosting permaneceu como o melhor modelo em CV, enquanto o FCNN manteve o melhor desempenho no conjunto de teste. Maiores de K aumentam consideravelmente o custo computacional sem trazer ganhos relevantes neste caso. O tempo de execução 32 segundos (K=5) para 1 minuto e 15 segundos (K=10).

- No modelo Gradient Boosting, as variáveis mais influentes para prever a conversão foram `residence_time`, `c0` e `T`. Essas variáveis apresentaram forte impacto na performance do modelo, enquanto as demais contribuíram minimamente, indicando dependência de poucos preditores dominantes.

- Já o modelo Random Forest distribuiu sua importância por um conjunto mais amplo de variáveis, incluindo `T`, `residence_time`, `log_residence_time`, `c0` e `mass_flux_density`. Isso sugere que o Random Forest capturou interações mais complexas e diversas nos dados, enquanto o Gradient Boosting se concentrou nos preditores mais fortes, o que pode limitar sua capacidade de generalização frente a distribuições ligeiramente diferentes.
